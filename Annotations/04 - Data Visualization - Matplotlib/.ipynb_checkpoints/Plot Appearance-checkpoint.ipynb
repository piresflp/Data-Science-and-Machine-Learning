```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=4, alpha=0.5, ls='--') 
# lw = linewidth
# alpha = how transparent the line is
# ls = linestyle
```
