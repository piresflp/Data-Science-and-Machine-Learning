* A single perceptron won't be enough to learn complicated systems.
* Fortunately, we can expand on the idea of a single perceptron, to create a multi-layer perceptron model. (basic neuron network)

* To build a network of perceptrons, we can connect layers of perceptron using a **multi-layer perceptron model**.

#### The outputs of one perceptron are directly fed into as inputs to other perceptron.




* The first layer: **input layer**.
* The last layer: **output layer**.
* Layers in between the input and output layers are the **hidden layers**.
* Hidden layers are difficult to interpret, due to their high interconnectivity and distance away from known input or output values.


* Neural Networks become <strong>"deep neural networks"</strong> if then contain 2 or more hidden layers.
