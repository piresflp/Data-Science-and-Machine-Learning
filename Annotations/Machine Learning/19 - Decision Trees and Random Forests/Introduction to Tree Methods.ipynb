

* **Nodes:**
    Split for the value of a certain attribute

* **Edges:**
    Outcome of a split to the next node.

* **Root:**
    The node that performs the first split.

* **Leaves:**
    Terminal nodes that predict the outcome.
    
    
### Random Forests    
To improve performance, we can use many trees with a random sample of features chosen as the split.
* A new random sample of features is chosen for <strong>every single tree at every single split.</strong>
* For <strong>classification</strong>, m is typically chosen to be the square root of p.
